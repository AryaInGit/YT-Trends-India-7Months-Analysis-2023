### Time-Based Analysis

```python
# Define the date range from January to July of the same year
start_date = datetime(2023, 1, 1)
end_date = datetime(2023, 7, 31)
```

Date Range Definition:
   - Defines a start date as January 1, 2023, and an end date as July 31, 2023.

```python
# Filter for the duration of interest (January to July)
duration_condition = (in_trending_videos['trending_date'] >= start_date) & (in_trending_videos['trending_date'] <= end_date)
in_trending_videos_duration = in_trending_videos[duration_condition]
```

Data Filtering:
   - Creates a boolean condition (`duration_condition`) to filter rows based on 'trending_date'.
   - Filters for rows where 'trending_date' falls between January and July 2023.
   - The filtered result is stored in `in_trending_videos_duration`.

# Daily View Trends of Trending Videos (Jan-Jul)

```python
# Extract day and month from publication date
in_trending_videos_duration['publication_day'] = in_trending_videos_duration['published_day'].dt.day
in_trending_videos_duration['publication_month'] = in_trending_videos_duration['published_day'].dt.month
```

- Extracts the day and month from the 'published_day' column.
- Creates new columns 'publication_day' and 'publication_month' to store these values.
- Assumes the 'published_day' column contains datetime values.

```python
# Group data by month and day and sum up the view counts
daily_views = in_trending_videos_duration.groupby(['publication_month', 'publication_day'])['view_count'].sum()
```

- The code groups video data by the combination of publication month and day.
- It calculates the total view counts for videos published on each specific month and day.
- The result is stored in a `daily_views` Series with a multi-level index.
- This approach helps analyze view count trends based on publication dates.

```python
# Create a line plot with colors for each month
plt.figure(figsize=(14, 7))
colors = ['blue', 'orange', 'green', 'red', 'purple', 'brown', 'pink']
for month in range(1, 8):  # Assuming Jan-Jul
    month_data = daily_views[month]
    plt.plot(month_data.index, month_data.values, label=f'Month {month}', color=colors[month-1])
    
    
plt.title('Trending Videos Over Days (Jan-Jul)')
plt.xlabel('Day')
plt.ylabel('Total View Count')
plt.xticks(rotation=45)
plt.legend(title='Month')
plt.grid(True)
plt.tight_layout() 
plt.show()
```

- The graph shows the number of trending videos over days in the month of July.
- The number of trending videos in the months of January, February, and March were lower than in July. 
- This suggests that July was the month with the most trending videos on YouTube India in 2023.           
Overall, the graph shows that the number of trending videos on YouTube in India is increasing. This is likely due to the growing popularity of the platform in India.

# Top 10 Most Common Tags

```python
# Top Tags Analysis
from collections import Counter

all_tags = ' '.join(in_trending_videos_duration['tags']).split()
tag_counter = Counter(all_tags)
top_tags = tag_counter.most_common(10)

plt.figure(figsize=(12, 8))
sns.barplot(x=[tag[1] for tag in top_tags], y=[tag[0] for tag in top_tags], palette='rocket')
plt.title('Top 10 Most Common Tags')
plt.xlabel('Tag Count')
plt.ylabel('Tag')
plt.tight_layout()
plt.show()
```
